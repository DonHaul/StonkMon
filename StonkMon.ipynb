```python
def GetHistoricalData(tickers,period,interval):
    data = yf.download(tickers=tickers,period=period,interval=interval,prepos=False,threads=True)
    
    data.index.name="timest"
    #df = pd.melt(data['Close'],ignore_index=False,var_name=["Metric","Asset"])
    data['Close']
    df = pd.melt(data['Close'],ignore_index=False,var_name=["Asset"])

    df=df.dropna()
    
    df['Diff'] =0
    df['100Dollar'] =0


    
    for el in tickers:
        #print(el)
        
        if df[df['Asset']==el].empty:
            continue
            
        #foreach element, fetch last column, value
        last_val= df[df['Asset']==el].tail(1)['value'][0] #try df tail
        
        multiplier = 100/last_val

        #df[df['Asset']=='AAL']
        df.loc[df['Asset']==el, 'Diff']= df.loc[df['Asset']==el, 'value']-last_val
        df.loc[df['Asset']==el, '100Dollar']= df.loc[df['Asset']==el, 'value']*multiplier
        
    return df
    
```
